```python
import pandas as pd
import numpy as np

#listas de datos
notas = [6.5,4.0, 7.0, 4.3, 5.7]
print(notas)
print("la primera calificacion es:", notas[0])
#diccionarios python

notas2_Estudiantes = {
    "ana": 6.5,
    "luis": 4.5,
    "benja": 5.8
}
promedio = np.mean(list(notas2_Estudiantes.values()))

print(f"El promedio es: {promedio:.2f}")
datos = {
    "nombre": ["ana", "luis", "maria", "pedro"],
    "nota": [6.5, 7.0, 4.3, 5.1]
}

df = pd.DataFrame(datos)
df_promediobajo5 = df[df["nota"] < 5]
print(df_promediobajo5)

# idxmax() encuentra la POSICIÓN (el índice) de la nota más alta
indice_maximo = df["nota"].idxmax()

# .loc usa ese índice para extraer la FILA completa
estudiante_mas_alto = df.loc[indice_maximo]

estudiante_mas_alto
```

```python
data = {
    'Producto':['Laptop', 'Mouse', 'Monitor', 'Teclado'],
    'Precio': [750000, 20000, 50000, 35000],
    'Stock':[10, 20, 15, 25]
}
```

```python

df_productos = pd.DataFrame(data)
print("DataFrame de productos:")
print(df_productos)
```

```python


promedio_stock = np.mean(df_productos['Stock'])
print(f"El promedio de stock de los productos es: {promedio_stock:.2f}")
```

```python
# 7. Filtrar los productos con un stock igual o mayor a 20.
productos_stock_mayor_igual_20 = df_productos[df_productos['Stock'] >= 20]
print("Productos con stock igual o mayor a 20:")
print(productos_stock_mayor_igual_20)
```

```python
print("\n" + "*"*30 + "\n")
```

```python
# 8. Mostrar el promedio del precio
promedio_precio = df_productos['Precio'].mean()
print(f"El promedio del precio de los productos es: {promedio_precio:.2f}")
```

```python
print("\n" + "*"*30 + "\n")
```

```python
# 9. Mostrar la cantidad total de todos los productos (suma de stock)
cantidad_total_stock = df_productos['Stock'].sum()
print(f"La cantidad total de stock de todos los productos es: {cantidad_total_stock}")
```

```python
print("\n" + "*"*30 + "\n")
```

```python
# 10. Mostrar la información de los productos con precio mayor o igual a 500000
productos_precio_mayor_igual_500k = df_productos[df_productos['Precio'] >= 500000]
print("Productos con precio mayor o igual a 500000:")
print(productos_precio_mayor_igual_500k)
```
